```python
import csv
import re

input_filename = 'gemini-code-1783951914382.txt'
output_filename = 'medicine_data.csv'

def convert_to_csv():
    try:
        # เปิดไฟล์ต้นทางเพื่ออ่านข้อมูล
        with open(input_filename, 'r', encoding='utf-8') as infile:
            raw_text = infile.read()

        # คลีนข้อมูล: ลบแท็ก ออกเพื่อให้ไฟล์ CSV สะอาด
        cleaned_text = re.sub(r'\\s*', '', raw_text)

        # ใช้ csv.reader อ่านข้อมูลจาก text ที่คลีนแล้ว 
        # (โมดูล csv จะจัดการเรื่องการขึ้นบรรทัดใหม่ภายใน "..." ให้เอง)
        csv_reader = csv.reader(cleaned_text.splitlines(), delimiter=',', quotechar='"')

        # เปิดไฟล์ปลายทางเพื่อเขียนข้อมูล (ใช้ utf-8-sig เพื่อให้อ่านภาษาไทยใน Excel ได้ไม่เพี้ยน)
        with open(output_filename, 'w', encoding='utf-8-sig', newline='') as outfile:
            csv_writer = csv.writer(outfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_ALL)
            
            # เขียนข้อมูลลงไฟล์ทีละแถว
            for row in csv_reader:
                # ข้ามแถวที่ว่างเปล่า
                if row:
                    # คลีนช่องว่างส่วนเกินหน้า/หลังข้อความในแต่ละเซลล์
                    cleaned_row = [cell.strip() for cell in row]
                    csv_writer.writerow(cleaned_row)

        print(f"แปลงไฟล์สำเร็จ! บันทึกข้อมูลลงในไฟล์: {output_filename}")

    except FileNotFoundError:
        print(f"ข้อผิดพลาด: ไม่พบไฟล์ชื่อ '{input_filename}' โปรดตรวจสอบชื่อไฟล์และตำแหน่งโฟลเดอร์")
    except Exception as e:
        print(f"เกิดข้อผิดพลาดขึ้น: {e}")

# เรียกใช้งานฟังก์ชัน
convert_to_csv()
```
